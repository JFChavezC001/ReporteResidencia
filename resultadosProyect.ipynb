## Superficie sembrada del año 2000

```python
import pandas as pd
```

```python
#creamos nuestras listas de datos que posteriormente transformaremos en un solo dataframe
FrijolPV=[1,"Frijol","primavera-verano","FPV",741,21217.5]
FrijolOI=[2,"Frijol","otoño-invierno","FOI", 168, ]
MFPV=[3,"Maíz Forrajero","primavera-verano","MFPV",7288,15395]
MFOI=[4,"Maíz Forrajero","otoño-invierno","MFOI",20,]
MGPV=[5,"Maíz de Grano","primavera-verano","MGPV",100612,487389]
MGOI=[6,"Maíz de Grano","otoño-invierno","MGOI",461,]
```

```python
supSembrada=[FrijolPV,FrijolOI,MFPV,MFOI,MGPV,MGOI]
```

```python
#Creamos el DF a traves de los datos de nuestras listas
supSemb=pd.DataFrame(supSembrada,columns=['Id','Cultivo','Ciclo productivo','Clave','Superficie Sembrada(ha)/Modalidad Riego','Superficie Sembrada(ha)/Modalidad Temporal'])
supSemb
```

```python
import matplotlib.pyplot as plt
```

```python
supSembradaR=plt.bar(supSemb['Clave'],
                    supSemb['Superficie Sembrada(ha)/Modalidad Riego'],
                   color='#158C2E')#agregamos los valores que se estan trabajando
plt.bar_label(supSembradaR)#agregamos las etiquetas con el valor de nuestra superficie para cada modalidad
plt.title("Superficie sembrada en la modalidad Riego para el Estado de México",loc="Center", pad= 20)#agregamos una etiqueta para nuestro titulo
plt.xlabel("Ciclo productivo")#agregamos etiqueta para los ciclos productivos en el eje X
plt.ylabel("Superficie sembrada (Hectareas)")#agregamos etiqueta para la superficie sembrada en el eje y
plt.show
```

```python
supSembradaT=plt.bar(supSemb['Clave'],
                    supSemb['Superficie Sembrada(ha)/Modalidad Temporal'],
                   color='#158C2E')#agregamos los valores que se estan trabajando
plt.bar_label(supSembradaT)#agregamos las etiquetas con el valor de nuestra superficie para cada modalidad
plt.title("Superficie sembrada en la modalidad temporal para el Estado de México",loc="Center",pad= 20)#agregamos una etiqueta para nuestro titulo
plt.xlabel("Ciclo productivo")#agregamos etiqueta para los ciclos productivos en el eje X
plt.ylabel("Superficie sembrada (Hectareas)")#agregamos etiqueta para la superficie sembrada en el eje y
plt.show
```

## Valor de Producción año 2000 

### Aplicamos las mismas opciones para el Valor de producción

```python
#Cargamos nuestros datos de valor de producción

FrijolPV=[1,"Frijol","primavera-verano","FPV",8155960,138886988]
FrijolOI=[2,"Frijol","otoño-invierno","FOI", 1541400, ]
MFPV=[3,"Maíz Forrajero","primavera-verano","MFPV",165246660,252315650]
MFOI=[4,"Maíz Forrajero","otoño-invierno","MFOI",229800,]
MGPV=[5,"Maíz de Grano","primavera-verano","MGPV",652819944.4,1859665490]
MGOI=[6,"Maíz de Grano","otoño-invierno","MGOI",3222700,]
```

```python
#Guardamos las listas en un solo diccionario
valpro=[FrijolPV,FrijolOI,MFPV,MFOI,MGPV,MGOI]
```

```python
ValProd=pd.DataFrame(valpro,columns=['Id','Cultivo','Ciclo productivo','Clave','Valor de Producción/Modalidad Riego','Valor de Producción/Modalidad Temporal'])
ValProd
```

```python
ValProdR=plt.bar(ValProd['Clave'],
                    ValProd['Valor de Producción/Modalidad Riego'],
                   color='#158C2E')#agregamos los valores que se estan trabajando
plt.ticklabel_format(style='plain', axis= 'y')  # Evita notación científica en el eje y
plt.bar_label(ValProdR,labels=[f'{int(v):,}' for v in ValProd['Valor de Producción/Modalidad Riego']]
)#agregamos las etiquetas con el valor de nuestra superficie para cada modalidad
plt.title("Valor de producción en la modalidad Riego para el Estado de México",loc="Center", pad= 20)#agregamos una etiqueta para nuestro titulo
plt.xlabel("Ciclo productivo")#agregamos etiqueta para los ciclos productivos en el eje X
plt.ylabel("Valor de producción")#agregamos etiqueta para la superficie sembrada en el eje y
plt.show
```

```python
ValProdT=plt.bar(ValProd['Clave'],
                    ValProd['Valor de Producción/Modalidad Temporal'],
                   color='#158C2E')#agregamos los valores que se estan trabajando
plt.ticklabel_format(style='plain', axis= 'y' )  # Evita notación científica en el eje y
plt.bar_label(ValProdT,labels=[f'{float(v):,}' for v in ValProd['Valor de Producción/Modalidad Temporal']]
)#agregamos las etiquetas con el valor de nuestra superficie para cada modalidad
plt.title("Valor de producción en la modalidad temporal para el Estado de México",loc="Center", pad= 20)#agregamos una etiqueta para nuestro titulo
plt.xlabel("Ciclo productivo")#agregamos etiqueta para los ciclos productivos en el eje X
plt.ylabel("Valor de producción")#agregamos etiqueta para la superficie sembrada en el eje y
plt.show
```
