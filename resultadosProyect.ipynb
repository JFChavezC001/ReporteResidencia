## Superficie sembrada del año 2000

```python
import pandas as pd
```

```python
#creamos nuestras listas de datos que posteriormente transformaremos en un solo dataframe
FrijolPV=[1,"Frijol","primavera-verano","FPV",741,21217.5]
FrijolOI=[2,"Frijol","otoño-invierno","FOI", 168, ]
MFPV=[3,"Maíz Forrajero","primavera-verano","MFPV",7288,15395]
MFOI=[4,"Maíz Forrajero","otoño-invierno","MFOI",20,]
MGPV=[5,"Maíz de Grano","primavera-verano","MGPV",100612,487389]
MGOI=[6,"Maíz de Grano","otoño-invierno","MGOI",461,]
```

```python
supSembrada=[FrijolPV,FrijolOI,MFPV,MFOI,MGPV,MGOI]
```

```python
#Creamos el DF a traves de los datos de nuestras listas
supSemb=pd.DataFrame(supSembrada,columns=['Id','Cultivo','Ciclo productivo','Clave','Superficie Sembrada(ha)/Modalidad Riego','Superficie Sembrada(ha)/Modalidad Temporal'])
supSemb
```

```python
import matplotlib.pyplot as plt
```

```python
supSembradaR=plt.bar(supSemb['Clave'],
                    supSemb['Superficie Sembrada(ha)/Modalidad Riego'],
                   color='#158C2E')#agregamos los valores que se estan trabajando
plt.bar_label(supSembradaR)#agregamos las etiquetas con el valor de nuestra superficie para cada modalidad
plt.show
```

```python
supSembradaT=plt.bar(supSemb['Clave'],
                    supSemb['Superficie Sembrada(ha)/Modalidad Temporal'],
                   color='#158C2E')#agregamos los valores que se estan trabajando
plt.bar_label(supSembradaT)#agregamos las etiquetas con el valor de nuestra superficie para cada modalidad
plt.show
```
